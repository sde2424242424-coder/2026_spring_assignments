```python
from faker import Faker
import pandas as pd
import random

fake = Faker('ko_KR')
Faker.seed(42)
random.seed(42)

class 사람:
    def __init__(self):
        self.이름 = fake.name()
        self.나이 = random.randint(18, 65)
        self.주소 = fake.address().replace("\n", " ")
        self.전화번호 = fake.phone_number()
        self.직업 = fake.job()
        self.이메일 = fake.email()

    def 정보출력(self):
        return f"{self.이름}, {self.나이}, {self.주소}, {self.전화번호}, {self.직업}, {self.이메일}"

    def 딕셔너리로변환(self):
        return {
            "이름": self.이름,
            "나이": self.나이,
            "주소": self.주소,
            "전화번호": self.전화번호,
            "직업": self.직업,
            "이메일": self.이메일
        }

사람데이터 = []

for _ in range(1000):
    한사람 = 사람()
    사람데이터.append(한사람.딕셔너리로변환())

df = pd.DataFrame(사람데이터)

print("===== 상위 5개 데이터 =====")
print(df.head())

print("\n===== 데이터프레임 정보 =====")
print(df.info())

print("\n===== 기본 통계 =====")
print(df.describe(include='all'))

df.to_csv("fake_people_1000.csv", index=False, encoding="utf-8-sig")
print("\n파일 저장 완료")
```

## **과제 2. Tom이 Jerry를 쫒아가는 행위를 method(함수)로 만들어 보자(10번 following)**

### **class User:**
> **def __init __ (self):**
>   - ##### Initialize attributes
>   - **HAS (Attiributes)**

> **def following_mode():**
>   - ##### change attributes (following, follower) for chasing
>   - **Does (Method, Change the attributes)**

<img src= 'https://pbs.twimg.com/profile_images/810136393/tom-jerry-pic_400x400.jpg' width=300>

```python
class User:
    def __init__(self, name):
        self.name = name
        self.following = 0
        self.followers = 0

    def following_mode(self, user):
        self.following += 1
        user.followers += 1

        print(f"{self.name}가 {user.name}를 팔로우합니다")
        print(f"{self.name}의 following: {self.following}")
        print(f"{user.name}의 followers: {user.followers}")


tom = User("Tom")
jerry = User("Jerry")

for i in range(10):
    tom.following_mode(jerry)
    tom.following_mode(jerry)
```
