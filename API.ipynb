4. Сравнение server header у Naver и Google
Код

Этот код сравнивает заголовки ответа серверов Naver и Google.

```python
import requests

urls = {
    "Naver": "https://www.naver.com",
    "Google": "https://www.google.com"
}

for name, url in urls.items():
    try:
        response = requests.get(url, timeout=10)

        print(f"\n{name}")
        print(f"URL: {url}")
        print(f"Status Code: {response.status_code}")

        server_header = response.headers.get("Server", "Server header not found")
        content_type = response.headers.get("Content-Type", "Content-Type not found")
        date_header = response.headers.get("Date", "Date header not found")

        print(f"Server: {server_header}")
        print(f"Content-Type: {content_type}")
        print(f"Date: {date_header}")

    except requests.exceptions.RequestException as e:
        print(f"\n{name}")
        print(f"Error: {e}")
```

6. Загрузка posts, выбор четных индексов и сохранение в Excel

Этот код загружает данные постов, выбирает строки с четным индексом и сохраняет их в Excel.

```python
import requests
import pandas as pd

url = "https://jsonplaceholder.typicode.com/posts"

response = requests.get(url)
data = response.json()

df = pd.DataFrame(data)

even_index_df = df[df.index % 2 == 0]

file_name = "even_index_posts.xlsx"
even_index_df.to_excel(file_name, index=False)

print("Even index rows:")
print(even_index_df)
print()
print(f"Excel file saved as: {file_name}")
```

7. Gradio UI: ввод userId и фильтрация posts в реальном времени

Этот код создает простой интерфейс на Gradio, в котором пользователь вводит userId, а программа показывает только подходящие данные.

```python
import requests
import pandas as pd
import gradio as gr

def get_posts(user_id):
    url = "https://jsonplaceholder.typicode.com/posts"

    params = {"userId": user_id}

    response = requests.get(url, params=params)
    data = response.json()

    df = pd.DataFrame(data)
    return df

iface = gr.Interface(
    fn=get_posts,
    inputs=gr.Number(label="Enter userId"),
    outputs=gr.Dataframe(label="Filtered Posts"),
    title="Posts Filter by userId",
    description="Enter a userId to filter posts in real time"
)

iface.launch()
```

8. Загрузка CSV, объединение и Gradio визуализация

Загружает CSV файл по ссылке и превращает его в DataFrame.

```python
import pandas as pd
import gradio as gr
import csv

def load_file(path):
    encodings = ["cp949", "euc-kr", "utf-8", "utf-8-sig"]

    for enc in encodings:
        try:
            df = pd.read_csv(
                path,
                encoding=enc,
                engine="python",
                on_bad_lines="skip",
                quoting=csv.QUOTE_MINIMAL
            )
            print(f"{path} loaded with {enc}")
            return df
        except:
            continue

    raise Exception(f"Cannot read file: {path}")


df1 = load_file("서울특별시 공공자전거 이용정보(일별)_2510.csv")
df2 = load_file("서울특별시 공공자전거 이용정보(일별)_2511.csv")
df3 = load_file("서울특별시 공공자전거 이용정보(일별)_2512.csv")


merged_df = pd.concat([df1, df2, df3], ignore_index=True)

print(merged_df.head())
print(merged_df.columns)

def show_data(n_rows):
    return merged_df.head(int(n_rows))

iface = gr.Interface(
    fn=show_data,
    inputs=gr.Number(label="Number of rows"),
    outputs=gr.Dataframe(label="Merged Data"),
    title="Seoul Bike Data Viewer"
)

iface.launch()
```

9. Gemini API / Agent AI (перевод через UI)

Функция принимает текст и возвращает перевод.

```python
import gradio as gr

def translate_text(text):

    translations = {
        "hello": "안녕하세요",
        "thank you": "감사합니다",
        "goodbye": "안녕히 가세요"
    }

    return translations.get(text.lower(), "Translation not found")

iface = gr.Interface(
    fn=translate_text,
    inputs=gr.Textbox(label="Enter English text"),
    outputs=gr.Textbox(label="Korean Translation"),
    title="Simple Translator (No API)",
    description="Translate basic English words to Korean"
)

iface.launch()
```

10. Hugging Face: попробовать мультимодальную LLM

Подключает pipeline() из библиотеки Transformers. Это стандартный способ быстро запускать готовые модели Hugging Face для инференса.

```python
import requests
from PIL import Image
from transformers import BlipProcessor, BlipForConditionalGeneration

model_name = "Salesforce/blip-image-captioning-base"

processor = BlipProcessor.from_pretrained(model_name)
model = BlipForConditionalGeneration.from_pretrained(model_name)

img_url = "https://huggingface.co/datasets/Narsil/image_dummy/raw/main/parrots.png"
raw_image = Image.open(requests.get(img_url, stream=True).raw).convert("RGB")

inputs = processor(raw_image, return_tensors="pt")
out = model.generate(**inputs)

caption = processor.decode(out[0], skip_special_tokens=True)
print(caption)
```
